```python
def splitStringDate(string):
    string = str(string)
    if len(string) == 10:
        ret = pd.to_datetime(f"{string[0:4]}.{string[4:6]}.{string[6:8]}.{string[8:10]}")
    if len(string) == 8:
        ret = pd.to_datetime(f"{string[0:4]}.{string[4:6]}.{string[6:8]}")
    return ret
```

```python
wetter_df = pd.read_csv("Wetter_Daten_Stündlich.csv",sep=";")
covid_df = pd.read_csv("covid_19-fallzahlen-kanton-stgallen-total.csv",sep=";")
wetter_df["date"] = wetter_df["time"].apply(splitStringDate)
```
